# Introduction

Visualization is a key topic for data science and data analysis. Because the human brain processes large amounts of complex data easier as images than as numbers in a spreadsheet, data visualization is incredibly important in understanding data and detecting patterns and trends. Fortunately, Python provides a lot of options in terms of executing visualizations for different purposes.

In the next two lessons, we will talk about the two most popular libraries for doing visualization in Python, `matplotlib` and `seaborn`. This lesson will focus on creating visualizations using `matplotlib` and the next lesson will focus on visual **exploratory data analysis with** `seaborn`.

Before we dive in, let’s go over some basic terminology.

![visualization](visualization.png)

Take a look at the above image, retreived from the `matplotlib` website. Let's explore some of the terms mentioned:

- `Figure`: This is the first top-level container. It is the overall window that contains everything that is drawn. A figure can have any number of Axes.
- `Axes`: This is the actual plot. It can have a lot of components to it, such as the x axis and the y axis, the plotting area, ticks, and so on. A figure can contain many Axes (known as subplots), but an Axes object can only be in one Figure.
- `Axis`: These are the number-line-like objects that set the plot limits and generate ticks.

## I. `pyplot` API

`matplotlib` is a plotting library that produces publication quality figures in a variety of formats and interactive environments. It provides an object-oriented interface, as well as a state-based interface. We will cover both in this lesson, starting with the state-based API.

The state-based interface `pyplot` is basically a collection of command style functions that make `matplotlib` work similar to MATLAB, if you are familiar with that language. The idea is that we have a collection of functions, and each function makes some changes to the current figure.

For instance, with one function we can create a figure, then with a second function we can create a plotting area in a figure. With a third function we can plot a line in a subplot of the figure, and then change the labels with a fourth function.

This will make more sense with an example. First, let’s import `matplotlib.pyplot`. The convention is to import this as `plt`:

Using the `plot` function, we can pass a list of numbers. When we execute the following, we will see that the command creates a figure.

```python
# Building a simple plot
## Plot a list of numbers
plt.plot([1, 2, 3, 2.5])
plt.show()
```

The figure contains one axes or subplot. Within this subplot we have a line graph that is just a graphical representation of the numbers we have in this list. This figure is considered to be the current figure. Every other function that we use will apply to this figure.

For instance, if we call the function `plt.ylabel`, it would place a label in the y axis (known as ylabel).

```python
# Building a simple plot
## Plot a list of numbers and label the y-axis
plt.plot([1, 2, 3, 2.5])
plt.ylabel('Some Numbers')
plt.show()
```

As you can see from this example, it is very convenient to use pyplot when working with just one subplot or one figure. But when we're working with more than one figure or more than one subplot in a single figure, it can get confusing. Let’s take a look at an example of this.

To get started, we will create some data:

```python
# Building a complex figure with two subplots
## Step 0: create data
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
```

The above snippet simply creates a function that takes an input t and multiplies the exponential of -t by the cosine of 2πt.

Now that we have data, let’s produce a figure with two subplots. First we create an empty figure using `plt.figure()`, then add subplots one at a time.

```python
# Building a complex figure with two subplots
## Step 1: create figure
plt.figure()

## Step 2: create subplot 1
plt.subplot(2, 1, 1)
plt.plot(t1, f(t1), 'bo')

## Step 3: create subplot 2
plt.subplot(2, 1, 2)
plt.plot(t2, f(t2), 'r--')

plt.show()
```

Let’s break this down. `plt.subplot` takes three integers, `nrows`, `ncols`, and `index`. We want a grid of two rows and one column, so nrows=2 and ncols=1. To create the first plot, we specify index=1. This plot took t1 as the x coordinates and f(t1) as the y coordinates. The ‘bo’ changes the plot from a simple line graph to a scatter plot with blue (the ‘b’ part) markers that are filled circles (the ‘o’ part).

Similarly, the second plot (index=2) takes t2 as the x coordinates and f(t2) as the y coordinates. The ‘r--’ changes the line to a red dashed line.

Now, what if we want to add a ylabel to the plot?

```python
# Building a complex figure with two subplots
## Step 1: create figure
plt.figure()

## Step 2: create subplot 1
plt.subplot(2, 1, 1)
plt.plot(t1, f(t1), 'bo')

## Step 3: create subplot 2
plt.subplot(2, 1, 2)
plt.plot(t2, f(t2), 'r--')

plt.ylabel("ylabel")   # which plot does this get added to?

plt.show()
```

The above snippet adds a ylabel to the second plot, but what if we want to add it to the first subplot? Or both subplots? We'd have to keep track of which subplot we are coding and the order in which we apply functions.

The state-based interface pyplot is useful for simple cases. But for more control and customization of plots, the object-oriented interface is preferable – we will cover this in the next step!

## II.  Object-oriented API

At its core, `matbplotlib` is object-oriented. To plot using the object-oriented API, typically you first create a figure, then call methods for an object within the figure to make changes to that object.

Let's look at a simple example using these arrays:

```python
# Simulate some data
x = np.linspace(-4, 4)
print(x)
y1 = x**2
y2 = x**3
```

The `plt.subplotsfunction` (not to be confused with the previous `plt.subplot` function) is used to create two objects at the same time – a figure object (usually called fig) and an axes object (usually called ax). We will produce a figure with one axes object, or one subplot.

```python
# Plot x, y1, y2 using object-oriented API
fig, ax = plt.subplots()
ax.plot(x, y1, 'r')
ax.plot(x, y2, 'b')
ax.set_title('Quadratic and Cubic Powers')
ax.set_xlabel('x values')
ax.set_ylabel('y values')

plt.show()
```

The above snippet creates fig and ax, then calls the plot method on the ax object to plot x and y1 as a red line and x and y2 as a blue line. Then, for the same object, we call `set_title`, `set_xlabel`, and `set_ylabel`.

Now, let's see how we can reproduce the same plot using the `pyplot` interface.

```python
# Plot x, y1, y2 using pyplot API
plt.plot(x, y1, 'r', x, y2, 'b')
plt.title('Quadratic and Cubic Powers')
plt.xlabel('x values')
plt.ylabel('y values')

plt.show()
```

We have fewer lines of code! We don’t need an explicit command to create a figure because the plot function will carry it out. We can also pass many arguments in these functions, so we can plot both lines at once.

So then what is the advantage of using the object-oriented interface? The advantage is in working with multiple subplots. Let’s take a look at an example.

In order to produce a grid of subplots, we again use the function plt.subplots, but this time we pass the number of rows and columns we want. Let’s produce a figure with four subplots:

```python
# Building a figure with four subplots
## Step 1: create grid
fig, axes = plt.subplots(nrows=2, ncols=2)
print(axes.shape)
```

Our object axes is a multi-dimensional NumPy array with four subplots. We can access each subplot using NumPy array notation. The first subplot (the upper-left one) has index 0, 0 so to make changes to this object, we just call the appropriate method!

```python
# Building a figure with four subplots
## Step 1: create grid
fig, axes = plt.subplots(nrows=2, ncols=2)

## Step 2: set title for each subplot
axes[0, 0].set_title('Upper Left')
axes[0, 1].set_title('Upper Right')
axes[1, 0].set_title('Lower Left')
axes[1, 1].set_title('Lower Right')

plt.show()
```

In the above snippet, we have called set_title to each of the four subplots. Now, what if we want to change all four objects at once? The nice thing about the object axes being a NumPy array is that we can easily iterate through every subplot using the flat attribute. Let’s set xticks and yticks as empty lists to remove them from all subplots at once:

```python
# Building a figure with four subplots
## Step 1: create grid
fig, axes = plt.subplots(nrows=2, ncols=2)

## Step 2: set title for each subplot
axes[0, 0].set_title('Upper Left')
axes[0, 1].set_title('Upper Right')
axes[1, 0].set_title('Lower Left')
axes[1, 1].set_title('Lower Right')

## Step 3: remove ticks from all subplots at once
for ax in axes.flat:
    ax.set(xticks=[], yticks=[])

fig.tight_layout()
plt.show()
```

We also called the method tight_layout on our figure object. This simply makes sure our subplots don’t overlap.

Now, let’s give our subplots some data! We will plot the first four powers of x in each of the subplots –

```python
# Building a figure with four subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

axes[0, 0].set_title('Linear')
axes[0, 0].plot(x, x)

axes[0, 1].set_title('Quadratic')
axes[0, 1].plot(x, x**2)

axes[1, 0].set_title('Cubic')
axes[1, 0].plot(x, x**3)

axes[1, 1].set_title('4th Power')
axes[1, 1].plot(x, x**4)

for ax in axes.flat:
    ax.set(xticks=[], yticks=[])

fig.tight_layout()
plt.show()
```

As you can see, by accessing each subplot and calling plot, we plotted one function in each subplot. Note that we increase the size of our figure using figsize.

Our figure is set to 10x5 inches, but Jupyter automatically minimizes large images to fit on the screen. If your screen is smaller than this, you can double-click on minimized images to increase the size and toggle scrollbars.

A more elegant approach is shown in the below example, which plots the first ten powers of x in a 2 x 5 grid.

```python
# A more elegant approach to builing figures with object-oriented API
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(15, 5))   # create 2x5 grid 
for i, ax in enumerate(axes.flatten()):                       # for each subplot,
    ax.set_title(f'x ** {i+1}')                                 #   set title to x**i (where i is the subplot number)
    ax.plot(x, x**(i+1))                                      #   plot x**i

fig.tight_layout()
plt.show()
```

This example clearly illustrates the convenience of using the object-oriented interface…we’ve plotted 10 subplots in just four lines of code!

## III. Customizations

Effective data visualization needs to be clear, concise, and tell a story. A visualization that is too plain may fail to capture the audiences’ attention, but a visualization that is too cluttered may be too difficult to understand. Fortunately matplotlib allows us to tweak every single element of a plot. In the previous two steps, we introduced a few customizations (such as line color and width). We will now delve deeper into these and other common customizations.

First, let's generate some data to work on

```python
# Generating data
x = np.linspace(-np.pi, np.pi, 200)
sine = np.sin(x)
cosine = np.cos(x)
```

#### Colors

The most common way whereby people use colors in matplotlib is according to initial or name. For every element, such as a bar, a line, or text, you can pass an additional argument with a string that contains the color that you want. If you want your element to be the color blue, you can pass the string b, or you can pass the whole word blue.

The following is a list of some commonly used colors and their initials:

`b`: blue
`g`: green
`r`: red
`c`: cyan
`m`: magenta
`y`: yellow
`k`: black
`w`: white

Other colornames that are allowed are the `HTML/CSS` colornames, such as burlywood and chartreuse. There is a list of 147 such colors; you can see them here.

Another way you can specify colors in matplotlib is using hexadecimal values. So, you can pass codes such as `#0000FF` and `matplotlib` will understand.

A gray level can be given instead of color by passing a string representation of a number between 0 and 1, where 0 is black and 1 is white. Every value in-between would be some shade of gray; for example, 0.75 would be a lighter shade of gray.

You can also pass RGBA tuples, where the first three values define the RGB color and the last (optional) value will be the transparency value, alpha.

The below is an example of all five methods:

```python
# Colors
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sine, color='r')                     # red
ax.plot(x, cosine, color='#165181')             # blue
ax.plot(x, x, color='0.75')                     # grey
ax.plot(x, -x, color='Indigo')                  # indigo
ax.plot(x, x**2, color=(0.4, 0.7, 0.3, 0.2))    # transparent green

plt.show()
```

#### Setting axis limits

Using the `pyplot API`, you can change axis limits using `.xlim()` and `.ylim()`. Using the object-oriented API, you can change the x axis limit with `set_xlim()` and the y axis limit with `set_ylim`. These methods take two values, the minimum and the maximum.

```python
# Axis limits
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sine, color='r')
ax.plot(x, cosine, color='#165181')

ax.set_xlim(-5, 5)
ax.set_ylim(-1.5, 1.5)

plt.show()
```

We see that the default limits for both of our axes are changed.

#### Setting ticks and tick labels
Ticks are the marks that you can see on the axis and the tick labels are the values that correspond to the tick marks. We can set both as needed. Let’s change our x ticks:

```python
# Tick marks
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sine, color='r')
ax.plot(x, cosine, color='#165181')

ax.set_xlim(-5, 5)
ax.set_ylim(-1.5, 1.5)

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])

plt.show()
```

This set our tick mark on the x axis. Now we can also change the tick labels…let’s change ours to show in terms of pi –

```python
# Tick labels
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sine, color='r')
ax.plot(x, cosine, color='#165181')

ax.set_xlim(-5, 5)
ax.set_ylim(-1.5, 1.5)

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'],
                  size=17)

plt.show()
```

#### Legend

The legend is used to identify the curves in your plot. To create a legend, we use the `legend()` method with an optional argument where you tell matplotlibwhere you want the label to be placed:

```python
# Legend
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sine, color='r', label='Sine')             # added label Sine
ax.plot(x, cosine, color='#165181', label='Cosine')   # added label Cosine

ax.set_xlim(-5, 5)
ax.set_ylim(-1.5, 1.5)

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'],
                  size=17)

ax.legend(loc='upper left')

plt.show()
```

Now we see the label for the sine and cosine functions placed on the upper left of the graph!

#### Annotations

There are two main methods for doing annotations in your plot. You can use the `text()` method, which takes the x coordinate and the y coordinate where you want your text to be shown as the first two arguments, and the actual text as the third argument.

```python
# Annotation
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sine, color='r', label='Sine')
ax.plot(x, cosine, color='#165181', label='Cosine')

ax.set_xlim(-5, 5)
ax.set_ylim(-1.5, 1.5)

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'],
                  size=17)

ax.legend(loc='upper left')

## Text method
ax.text(-0.25, 0, "(0, 0)")                         # label (0, 0)
ax.text(np.pi - 0.25, 0, r'$(\pi, 0)$', size=15)    # label (pi, 0)

plt.show()
```

Or, you can use the `annotate()` method. The location being annotated is specified in the argument xy, and the position of the annotation text is specified by the argument xytext.

```python
# Annotation
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sine, color='r', label='Sine')
ax.plot(x, cosine, color='#165181', label='Cosine')

ax.set_xlim(-5, 5)
ax.set_ylim(-1.5, 1.5)

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'],
                  size=17)

ax.legend(loc='upper left')

ax.text(-0.25, 0, "(0, 0)")                       # label (0, 0)
ax.text(np.pi - 0.25, 0, r'$(\pi, 0)$', size=15)  # label (pi, 0)

## Annotate method
ax.annotate('Origin',                             # text
            xy=(0, 0),                            # arrow point
            xytext=(1, -0.7),                     # locaton of text
            arrowprops=dict(facecolor='blue'))    # makes arrow blue

plt.show()
```

Let's break down the above snippets. In the first, we labeled the points (0, 0) and (π, 0) using `text()`. In the second, using `annotate()`, we annotated the point (0, 0) with an arrow who’s tip starts at that point (xy argument). The end of this arrow is at the point (1, -0.7) (xytext argument) displays the text “Origin”.

We passed a dictionary for the arrow. This dictionary has many optional keyword arguments – you can find out more here.

#### Producing grids, horizontal, and vertical lines
There are a couple of methods for producing grids, horizontal, and vertical lines. The `axhline()` method will produce horizontal lines and you can specify other arguments, such as alpha and color. Similarly, we have the `axvline()` method that plots vertical lines. The `grid()` method simply displays the grid for your plot.

Let’s take a look at all three of these –

```python
# Grids and lines
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sine, color='r', label='Sine')
ax.plot(x, cosine, color='#165181', label='Cosine')

ax.set_xlim(-5, 5)
ax.set_ylim(-1.5, 1.5)

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'],
                  size=17)

ax.legend(loc='upper left')

ax.text(-0.25, 0, "(0, 0)")
ax.text(np.pi - 0.25, 0, r'$(\pi, 0)$', size=15)

ax.annotate('Origin',
            xy=(0, 0),
            xytext=(1, -0.7),
            arrowprops=dict(facecolor='blue'))

ax.axhline(0, color='black', alpha=0.9)   # horizontal line at x=0
ax.axvline(0, color='black', alpha=0.9)   # vertical line at y=0
ax.grid()                                 # shows grid

plt.show()
```

You can customize almost every element in matplotlib, but the ones that we have undertaken are the commonly used changes that you will want to do.
